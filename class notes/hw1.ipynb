### 1. Create a function to_co2eq(tons) that returns kilograms of CO2eq (10 pts)

```python
def to_co2eq(tons): # im assuming I need to convert us tons to kilograms ( ill make a metric conversion just in case)
    return tons * 907.185

def mt_co2eq(mtons):
    return mtons * 1000

# now im going to check to see if the equations work.
t = 10
a = to_co2eq(t)

mt = 10
b = mt_co2eq(mt)

print(t,"tons to kilograms is", (str(round(a, 2))+"kg"))
print(mt,"metric tons to kilograms is", (str(b)+"kg"))
```

### 3. Add a new dam to dams.csv (in the lecture note) and re-run the pandas cells to see it appear (20 pts)

```python
import csv
csv_path = "dams.csv"
rows = [
    {"name": "DamA", "capacity_MW": 120, "country": "BR"},
    {'name': 'DamB', 'capacity_MW': 300, 'country': "PE"},
    {'name': 'DamC', 'capacity_MW': 90, 'country': "CO"},
    {'name': 'DamD', 'capacity_MW': 100, 'country': 'US'}
]

with open(csv_path, "w", newline="", encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=["name", "capacity_MW", "country"])
    writer.writeheader()
    writer.writerows(rows)

with open(csv_path, "r", encoding='utf-8') as f:
    print(f.read())
```

### 5. Modify the plotting cell to display a line plot of cumulative capacity (20pts).


```python
import matplotlib.pyplot as plt

cumulative_cap = df["capacity_MW"].cumsum()
labels = df["name"].tolist()

plt.figure()
plt.plot(labels, cumulative_cap, marker='o')
plt.title("Cumulative Installed Capacity by dam (MW)")
plt.xlabel("Dam")
plt.ylabel("Capacity (MW)")
plt.show()
```

# 7. Exceptions
handle errors gracefully using try/except.

```python
def safe_divide(a, b):
    try:
        return a/b
    except ZeroDivisionError as e:
        return f"Error: {e}"

safe_divide(10, 0)
```

# 8. Sets
A set is a container which contains unordered collections of unique itrems. They are defined like lists but they use curly brackets as the dictionaries.


```python
primes = {2, 3, 5, 7}
evens = {2, 4, 6, 8, 10}
odds = {1, 3, 5, 7, 9}

print(type(primes))
```

Sets supports well known algebraic set operations:

- union(*others)
- set | other | ...

return a new set with elements from the set and all others

```python
print(evens)
print(odds)
allnumbers = evens | odds
allnumbers
```

- Intersection(*others)
- set & other & ...

Return a new set with elements common to the set and all otehrs

```python
primes & odds
```

- difference(*others)
- set-other-...

Return a new set with elements in the set that are not in the others

```python
evens - primes
```

- symmetric_difference(other)
- set^other

return a new set with elements in either the set or other but not both

```python
primes^odds
```

- .isdisjoint(other)

Return Ture if the set has no elements in common with other. Sets are disjoint if and only if their intersection is the empty set

```python
evens.isdisjoint(odds)
```

- .issubset(other) set <= other

tests whether every element in the set is in other

```python
primes.issubset(allnumbers)
```

- set < other

tests whether the set is a proper subset of other, that is set<= other and set != other

```python
A = {1,2,3}
B = {1,2,3,4}

A < B
```

- .issuperset(other)

Test whether every element in other is in the set

```python
print(primes.issuperset(odds))
allnumbers.issuperset(odds)
```

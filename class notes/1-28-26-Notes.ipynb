 ## Loops: for, while, and comprehensions

```python
capacities = [75, 100, 215, 300, -5] # example data
bins = [0, 100, 200, 300]

for cap in capacities:
    for i in range(len(bins)-1):
        if bins[i] <= cap < bins[i + 1]:
            print(f"{cap} MW is in [{bins[i], bins[i+1]})")
            break
    else:
        print(f"{cap} MW is out of range.")

count = 3
while count > 0:
    print("countdown:", count)
    count -= 1

sq = [n*n for n in range(6)]
print("Squares:", sq)
```

## Functions and scope
use functions to encapsulate reusable logic.

```python
# A simple function with no arguments
def say_hello():
    return "Hello, World!"

print(say_hello())

# a function with parameters
def add_numbers(a,b):
    return a + b

print("Sum:", add_numbers(5,7))

# a function with a default parameter
def greet(name="Student"):
    return f"Hello, {name}!"

print(greet())
print(greet("Hydropower Researcher"))

# a function with multiple return values
def min_max(numbers):
    return min(numbers), max(numbers)

values = [3, 7, 2, 9, 5]
low, high = min_max(values)
print(f"Lowest: {low}, Highest: {high}")
```

## Exceptions
handle errors gracefully using try/except.

```python
def safe_divide(a,b):
    try:
        return a/b
    except ZeroDivisionError as e:
        return f"Error: {e}"

safe_divide(10, 0)
```

## Files: read and write (tex, CSV, JSON)
python's built in open handles text files; csv and json handle structred data

```python
# 8.1 text file
text_path = "notes.txt"
with open(text_path, "w", encoding="utf-8") as f:
    f.write("Hydropower notes\n- Example line 1\n- Example line 2")

with open(text_path, "r", encoding='utf-8') as f:
    print(f.read())

# 8.2 CSV using csv module
import csv
csv_path = "dams.csv"
rows = [
    {"name": "DamA", "capacity_MW": 120, "country": "BR"},
    {'name': 'DamB', 'capacity_MW': 300, 'country': "PE"},
    {'name': 'DamC', 'capacity_MW': 90, 'country': "CO"},
]

with open(csv_path, "w", newline="", encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=["name", "capacity_MW", "country"])
    writer.writeheader()
    writer.writerows(rows)

with open(csv_path, "r", encoding='utf-8') as f:
    print(f.read())

# 8.3 JSON
import json
json_path = "dam.json"
with open(json_path, "w", encoding='utf-8') as f:
    json.dump(rows, f, indent=2)

with open(json_path, "r", encoding='utf-8') as f:
    print(f.read())
```

## Tabular data with pandas and DataFrames
pandas is the standard library for data manipulation. Lets read the csv we wrote and do basic operations

```python
import pandas as pd
print(pd.__version__)
df = pd.read_csv("dams.csv")
print(type(df))
df
```

```python
# basic exploration
df.info()
df.describe()
```

```python
# Filtering, sorting, new columns
high_cap = df[df["capacity_MW"] >= 100].sort_values("capacity_MW", ascending=False)
high_cap["power_density"] = high_cap["capacity_MW"] / 1.0  # pretend flooded area = 1 km^2
high_cap
```

## simple ploting wth matplotlib
when plotting, well use matplotlib directly and avoid specifying colors per instructions.

```python
import matplotlib.pyplot as plt

cumulative_cap = df["capacity_MW"].cumsum()
labels = df["name"].tolist()

plt.figure()
plt.plot(labels, cumulative_cap, marker='o')
plt.title("Cumulative Installed Capacity by dam (MW)")
plt.xlabel("Dam")
plt.ylabel("Capacity (MW)")
plt.show()
```
